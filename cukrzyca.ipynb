```python
from sklearn.datasets import load_diabetes

diabetes_dataset = load_diabetes()
```

```python
import pandas

diabetes_dataframe = pandas.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
```

```python
pandas.plotting.scatter_matrix(diabetes_dataframe, figsize=(20,20), c = diabetes_dataset.target)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =train_test_split(diabetes_dataset.data, diabetes_dataset.target, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN_model = KNeighborsRegressor(n_neighbors=30,)
KNN_model.fit(X_train, y_train)
print(KNN_model.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)

forest_model.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPRegressor

neural_network = MLPRegressor()
neural_network.fit(X_train, y_train)

neural_network.score(X_test, y_test)
```

```python
from sklearn.feature_selection import VarianceThreshold

variance = VarianceThreshold()

X = diabetes_dataset.data.copy()

variance.fit_transform(X)
```

```python
forest_model2 = RandomForestRegressor()

forest_model2.fit(X_train, y_train)

forest_model2.score(X_test, y_test)
```
